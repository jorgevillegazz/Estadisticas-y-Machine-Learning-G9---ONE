Sin embargo, tenemos **otro método** que nos permite ver mejor.

¿Cómo lo conseguimos?Aquí de nuevo tenemos una documentación que podemos usar, que es utilizando el **crosstab**.¿Y qué significa el crosstab?

**La función crosstab en pandas** sirve para **calcular tablas de frecuencia cruzada (contingencia) de dos o más factores**, lo que permite analizar relaciones, patrones y tendencias entre variables categóricas. Genera un DataFrame que resume cómo se distribuyen los datos, mostrando frecuencias absolutas o porcentajes.

**Principales usos y características:**

*Análisis de frecuencia*: **Por defecto**, cuenta cuántas veces ocurren las combinaciones de categorías (frecuencia).

*Tabla de Contingencia*: Ideal para cruzar datos como encuestas o estudios de mercado.

*Normalización*: Con normalize=True, **calcula porcentajes** en lugar de conteos absolutos.

*Márgenes:* Permite **añadir filas/columnas de totales** usando margins=True.

*Agregación:* Puede **calcular estadísticas (media, suma, etc.)** sobre una tercera variable numérica usando values y aggfunc.


Nosotros podemos, de alguna manera, ver dos variables al mismo tiempo .Por ejemplo, estamos viendo color y sexo, o podríamos ver otra variable que también sea importante.solo que creé un diccionario que clasifica masculino y femenino para sexo, y para color con categorías como indígena, blanco, negro, amarillo, etc.Entonces, ¿cómo empezamos a usar el método del crosstab?

```python
# Creamos  un par de  Dicionarios, solo para llegar a cambiar el nombre   mas adelante de la  tabla
sexo = {0: 'Masculino',
        1: 'Femenino'}

color = {0: 'Indigena',
         2: 'Blanco',
         4: 'Negro',
         6: 'Amarillo',
         8: 'Pardo',
         9: 'Sin declarar'}
```

```python
#Creamo una  nueva variable(tabla) y ...
# Extraemos los datos de la tabla(datos) en este caso los de las  columnas  Sexo y Color
frecuencia_cross =pd.crosstab(datos.Sexo, datos.Color)
frecuencia_cross
```

Una vez  ejecutado  apreciamos  la nueba tabla con los datos " CRUZADOS"

Solo que tanto los  nombres de indices y columnas un usiario no los podría interpretar.

Es por eso  que  les cambiaremos  los nombres  para facilitar su entendimiento.

```python
# Ahora  si  usaremos  los dos diccionarios que creamos anteriormente (sexo y color)
frecuencia_cross.rename(index=sexo,inplace=True)
frecuencia_cross.rename(columns=color,inplace=True)
frecuencia_cross
```

Supongamos que  quiero ver también cómo se ve la proporción, además de los valores absolutos, cómo hacer?.

```python
porcentaje2 = pd.crosstab(datos.Sexo,
                         datos.Color,
                         normalize=True).round(4)*100

porcentaje2.rename(index = sexo,inplace = True)
porcentaje2.rename(columns = color,inplace = True)
porcentaje2
```

En el mundo del análisis de datos, a menudo es necesario explorar la relación entre dos variables categóricas. El código que hemos visto utiliza la función pd.crosstab de Pandas para crear una tabla de contingencia(cruzada), que nos **permite examinar cómo se distribuyen los valores de una variable** (*Ingreso en este caso*) en función de las combinaciones de otras dos variables categóricas (Sexo y Color).

```python
porcentaje_ingreso = pd.crosstab(datos.Sexo,
                         datos.Color,
                         aggfunc = 'mean',
                         values = datos.Ingreso)

porcentaje_ingreso.rename(index = sexo,inplace = True)
porcentaje_ingreso.rename(columns = color,inplace = True)
porcentaje_ingreso
```

# **Explicación paso a paso:**

* pd.crosstab(datos.Sexo, datos.Color...): Esta es la función principal que crea la tabla de contingencia.

* Los dos primeros argumentos (datos.Sexo y datos.Color) especifican las dos variables categóricas que se cruzarán.

* aggfunc = 'mean': Indica que queremos calcular la media de datos.Ingreso para cada combinación de Sexo y Color. Podríamos usar otras funciones de agregación como sum, count, etc., dependiendo del análisis que queramos realizar.

* values = datos.Ingreso: Especifica la variable numérica cuyos valores se agregarán en cada celda de la tabla.

* porcentaje.rename(index = sexo, inplace = True): Esta línea renombra las etiquetas del índice (filas) de la tabla de contingencia utilizando los valores del diccionario sexo. El argumento inplace = True modifica la tabla directamente en lugar de crear una nueva.

* porcentaje.rename(columns = color, inplace = True): De manera similar, esta línea renombra las etiquetas de las columnas de la tabla utilizando los valores del diccionario color.

* porcentaje: Finalmente, se muestra la tabla resultante.

# ¿Qué nos muestra esta tabla?

La tabla resultante nos mostrará ***la media de Ingreso para cada combinación de Sexo y Color***. Esto nos permite comparar, por ejemplo, si hay diferencias significativas en los ingresos promedio entre hombres y mujeres de diferentes etnias.

Hay que experimentar con diferentes funciones de agregación y variables en pd.crosstab **para descubrir patrones interesantes en tus datos**.

¡**El análisis bivariado** es una herramienta poderosa para comprender mejor las relaciones entre variables en tus conjuntos de datos! ;))

## <font color=#56B4E9>2.3 Distribución de frecuencia para variables cuantitativas (clases de amplitud fija)</font>
***

Ahora, tenemos esta distribución de frecuencias personalizadas cuantitativas. Si queremos algo que tenga una amplitud fija, existe una teoría que se llama la <font color = green>**regla de Sturges**</font>.

Para eso, ¿qué es lo que sugiere? Imaginen que están organizando su biblioteca en casa, solo que tienen una inmensa cantidad de libros, y necesitan saber cuántos estantes van a querer para su biblioteca. Para saber este número de estantes, hacen la regla de Sturges. Él va a definir correctamente un estante donde tus libros no queden ni muy apretados ni estén danzando ahí en su estante. Básicamente, lo que hace es este cálculo

***

La ***regla de Sturges*** es un criterio estadístico utilizado para determinar el número óptimo de intervalos o clases (**k**) necesarios para organizar un conjunto de datos en un histograma o tabla de frecuencias. Propuesta por Herbert Sturges en 1926, equilibra el detalle de los datos con su interpretabilidad

### Importando bibliotecas

http://www.numpy.org/

```python
import numpy as np
```

### Paso 1 - Difiniendo el número de clases

#### Regla de Sturges

# $$k = 1 + \frac {10}{3}\log_{10}n$$

 n= (Total de observaciones): Es el tamaño de la muestra.

 log _10: Se utiliza el logaritmo común (base 10).

 k (Número de clases): Si el resultado no es entero, generalmente se redondea al número entero más cercano o al siguiente número impar para facilitar la visualización.



Primero tenemos  que ver el tamaño de nuestra base de datos

```python
datos.shape
```

**Sustituimos los datos de la formula**

Entonces n= el tamaño de los registros o **tamaño de la mustra**


```python
n=datos.shape[0] #[0] porque no  me interesa las columnas
```

Ahora nos toca  calcular logaritmo de 10 entre  3 x n  mas 1


```python
k=1+(10/3)*np.log10(n)
```

```python
k
```

k nos está mandando a hacer 17 clases.

Entonces, para hacer estas 17 clases,

Entonces, si él nos mandó a hacer 17, ¿cómo creamos esa tabla de frecuencia?

### Paso 2 - Crear la tabla de frecuencias

```python
frec_sturges = pd.value_counts(
                                pd.cut(x = datos['Ingreso'],
                                 bins=17,
                                 ##labels=label,
                                include_lowest=True))
```

```python
frec_sturges
```

Tenemos aquí Ingreso.¿Cuánta es la frecuencia que nosotros tenemos?

Básicamente, nos está mostrando, (*no se preocupen que aquí está en negativo,porque es de 200.000 a 11.764*), 11.764 y por delante.

Entonces, él está haciendo de manera mucho más detallada lo que nosotros hemos hecho de manera personalizada.

Esto es una manera también que se puede usar.Sin embargo, si ustedes ya conocen su base de datos,ustedes pueden hacer algo más personalizado para que no quede en este entendimiento.

***

Ahora  como ejersicio Haemos  la  tabla de sturges con  proporciones(porcentajes)

```python
porcentaje_sturges = pd.value_counts(
                                pd.cut(x = datos['Ingreso'],
                                      bins=17,
                                      include_lowest=True),
                                sort=False,
                                normalize=True).round(4)*100
```

```python
porcentaje_sturges
```

Bien, ahora formaremos  nustra nueva tabla de **Distribución de frecuencia para variables cuantitativas** (clases de amplitud fija)

```python
dist_frec_cuantitativa_amp_fija = pd.DataFrame({'Frecuencia':frec_sturges,'Porcentaje %':porcentaje_sturges})
```

```python
dist_frec_cuantitativa_amp_fija
```

## <font color=#56B4E9>2.4 Histograma</font>
***

El <b>HISTOGRAMA</b> es la representación gráfica de una distribución de frecuencia. Es un gráfico formado por un conjunto de rectángulos colocados uno al lado del otro, donde el área de cada rectángulo es proporcional a la frecuencia de la clase que representa.

## <font color=#56B4E9>3.2 Mediana</font>
***

Para obtener la mediana de un conjunto de datos, debemos proceder de la siguiente manera:
1. Ordenar el conjunto de datos;
2. Identificar el número de observaciones (registros) del conjunto de datos ($n$);
3. Identicar el elemento del medio:

> Cuando $n$ sea impar, la posición del elemento del medio se obtendrá de la siguiente manera:


# $$Elemento_{Md} = \frac{n+1}{2}$$

> Cuando $n$ sea par, la posición del elemento mediano se obtendrá de la siguiente manera:


# $$Elemento_{Md} = \frac{n}2$$

4. Obtener la mediana:

> Cuando $n$ sea impar:


# $$Md = X_{Elemento_{Md}}$$

> Cuando $n$ sea par:


# $$Md = \frac{X_{Elemento_{Md}} + X_{Elemento_{Md}+1}}2$$

Las **medidas de localización** son exactamente herramientas estadísticas que nos permiten entender cómo está distribuido nuestro conjunto de datos y nos proporcionan un análisis un poco más de valores relativos, diferentes de unas medidas de tendencia central, donde nosotros podemos ver el comportamiento y cómo tiende a estar nuestro conjunto de datos

***

Las medidas de localización son herramientas estadísticas que dividen un conjunto de datos ordenados en partes iguales para analizar la posición de valores específicos dentro de una distribución. Sirven para interpretar, comparar y entender la tendencia central o la dispersión relativa de los datos, siendo los más comunes cuartiles, deciles, percentiles y la mediana.

***

Principales Medidas de Localización y Ejemplos

* **Mediana (Q2 / P50)**: Divide los datos en dos partes iguales (50% por debajo y 50% por encima).Ejemplo: En el conjunto ordenado \(\{3,5,7,9,11\}\), la mediana es 7.


* **Cuartiles (Q1, Q2, Q3)**: Dividen la muestra en cuatro partes iguales (25% cada una).Ejemplo: Si el Q1 (25%) de calificaciones es 60, significa que el 25% de los estudiantes obtuvo 60 o menos.

Entonces, eso es súper importante que nosotros interpretemos.

La **frecuencia acumulativa** es una forma de **mostrar cuántos datos se han acumulado hasta un cierto punto** en un conjunto de datos. En lugar de simplemente contar cuántos valores hay en cada categoría o intervalo (como en una frecuencia simple), la frecuencia acumulativa suma las frecuencias de todas las categorías o intervalos anteriores.

Para entenderlo mejor, imagina que tienes las calificaciones de un grupo de estudiantes en un examen:

Categorías/Intervalos: Primero, divides las calificaciones en rangos (por ejemplo, 0-5, 6-7, 8-9, 10).

Frecuencia Simple: Luego, cuentas cuántos estudiantes hay en cada rango:

0-5: 2 estudiantes

6-7: 5 estudiantes

8-9: 8 estudiantes

10: 3 estudiantes

Frecuencia Acumulativa: Ahora, calculas la frecuencia acumulativa sumando las frecuencias de cada rango y las de los rangos anteriores:

0-5: 2 estudiantes (acumulado: 2)

6-7: 5 estudiantes (acumulado: 2 + 5 = 7)

8-9: 8 estudiantes (acumulado: 7 + 8 = 15)

10: 3 estudiantes (acumulado: 15 + 3 = 18)

En resumen, **la frecuencia acumulativa te dice cuántos estudiantes obtuvieron una calificación igual o inferior a un cierto valor**. En este caso, 15 estudiantes obtuvieron una calificación de 9 o menos.

## <font color=#56B4E9>4.2 Box-plot</font>
***

El *box plot* da una idea de la posición, dispersión, asimetría, colas y valores extremos. La posición central está dada por la mediana y la dispersión por $IIQ$ (distancia inter cuartil). Las posiciones relativas de $Q1$, $Mediana$ y $Q3$ dan una idea de la simetría de la distribución. Las longitudes de las colas están dadas por las líneas que van desde el rectángulo a los valores remotos y por los valores extremos.

Un **boxplot** (diagrama de caja y bigotes) en Python es una representación gráfica que resume la distribución, dispersión, simetría y valores atípicos (outliers) de un conjunto de datos numéricos a través de sus cuartiles (\(Q1\), mediana, \(Q3\)). Es ideal para análisis exploratorio de datos, permitiendo comparar distribuciones entre grupos

Para qué sirve un Boxplot en Python:

* Identificar valores atípicos: Resalta puntos fuera de los "bigotes" (outliers), facilitando la limpieza de datos.

* Visualizar la dispersión y asimetría: La caja representa el 50% central de los datos (Rango Intercuartílico - IQR), mostrando qué tan agrupados están.

* Comparar grupos: Seaborn o Matplotlib permiten crear **múltiples cajas para comparar** categorías fácilmente.

* Resumen estadístico rápido: Muestra simultáneamente la mediana, los cuartiles y el rango en una sola figura compacta.

Un boxplot típico consta de una caja (que va de \(Q1\) a \(Q3\)) con una línea central (mediana) y líneas (bigotes) que se extienden a los valores máximo y mínimo, exceptuando los valores atípicos marcados individualmente

## <font color=#56B4E9>5.2 Varianza</font>
***

### Varianza

La varianza se construye a partir de las diferencias entre cada observación y la media de los datos, es decir, la desviación alrededor de la media. Al calcular la varianza, las desviaciones alrededor de la media son elevadas al cuadrado.

### Varianza de la población

# $$\sigma^2 = \frac 1n\sum_{i=1}^{n}(X_i-\mu)^2$$

### Varianza de la muestra

# $$S^2 = \frac 1{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2$$

Continuando con la explicación de nuevas medidas de dispersión, vamos a abordar la **varianza y la desviación** estándar.

¿Qué es la varianza, por ejemplo? Es la diferencia entre cada observación y la media de los datos, es decir, la desviación que tenemos alrededor de la media.

Al calcular la desviación, debemos entender que existe un cálculo diferente para la varianza de la población y la varianza de la muestra. Usaremos la varianza de la muestra, ya que estamos trabajando con una base muestral.

Para ejemplificar el cálculo de la varianza, continuaremos usando nuestro DataFrame de notas de María. Las notas incluyen asignaturas, calificaciones, desviación y desviación absoluta.



```python
notas_Maria_dispersion
```

Queremos calcular la varianza, es decir, cuánto varía respecto a nuestra media. Para ello, crearemos una nueva columna llamada desviación elevada al cuadrado. Esto nos permitirá calcular la varianza.

```python
#la nueva columna se llamara desviaición elevada al cuadrado  que
#Es tomada  de la formula para la VARIANZA DE LA MUESTRA
notas_Maria_dispersion['(Desviacion)^2'] = notas_Maria_dispersion['Desviacion'].pow(2)
notas_Maria_dispersion
```

Esta nueva columna representa el cálculo de la desviación al cuadrado. Utilizaremos la función pow para elevar al cuadrado.

 Imprimiremos el resultado para verificar el cálculo correcto. Así obtenemos nuestra varianza.

Para calcular la varianza, sumamos la desviación elevada al cuadrado y la dividimos por el tamaño de la muestra menos uno. Usamos len para medir el tamaño.

```python
notas_Maria_dispersion['(Desviacion)^2'].sum() / (len(notas_Maria_dispersion) - 1)
```

Ahora tenemos la varianza, que es 4.5.

 Existe una manera más rápida de calcular esto de forma menos manual, utilizando la función var.

```python
notas_Maria_dispersion['María'].var()
```

## <font color=#56B4E9>5.3 Desviación estándar</font>
***

Una de las restricciones de la varianza es el hecho de que proporciona medidas cuadráticas de las unidades originales; la varianza de las medidas de longitud, por ejemplo, está en unidades de área. Por lo tanto, el hecho de que las unidades sean diferentes dificulta la comparación de la dispersión con las variables que la definen. Una forma de eliminar esta dificultad es considerar su raíz cuadrada.

### Desviación estándar de la población

# $$\sigma = \sqrt{\frac 1n\sum_{i=1}^{n}(X_i-\mu)^2} \Longrightarrow \sigma = \sqrt{\sigma^2}$$


### Desviación estándar de la muestra

# $$S = \sqrt{\frac 1{n-1}\sum_{i=1}^{n}(X_i-\bar{X})^2} \Longrightarrow S = \sqrt{S^2}$$

```python
varianza= notas_Maria_dispersion['María'].var()
```

```python
varianza
```

Finalmente, abordamos la desviación estándar, que es una medida cuadrática que proporciona unidades originales. Es importante tener cuidado al usar la varianza, ya que puede cambiar la interpretación de los datos, especialmente en variables como altura, donde elevar al cuadrado transforma metros en área.


Para calcular la desviación estándar, utilizamos numpy con la función sqrt.

```python
np.sqrt(varianza)
```

También podemos usar la función std de Python para calcular la desviación estándar de manera más rápida.

```python
desviacion_estandar = notas_Maria_dispersion['María'].std()
desviacion_estandar
```

La desviación estándar es 2.13. Aunque existen funciones rápidas, es importante entender el cálculo detrás de ellas.
